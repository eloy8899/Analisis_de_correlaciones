1.   Pearson [-1,1]

*   Mide relaciones lineales
*   Las variables tienen distribuci√≥n normal, se puede obviar cuando $n \to \infty$ (grande) +500
*   Es sensible a outliers
*   Mayor correlaci√≥n no implica causalidad

$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2} \sqrt{\sum (y_i - \bar{y})^2}}$

La f√≥rmula de la correlaci√≥n de Pearson mide cu√°nto se desv√≠an simult√°neamente las variables ùë•  e ùë¶ respecto a sus respectivas medias. Si ambas variables tienden a desviarse en el mismo sentido (es decir, cuando una aumenta, la otra tambi√©n), la correlaci√≥n ser√° positiva. Si se desv√≠an en sentidos opuestos (cuando una aumenta y la otra disminuye), la correlaci√≥n ser√° negativa.
Posteriormente, esta medida conjunta se normaliza dividi√©ndola entre los productos de las desviaciones est√°ndar de cada variable, lo que garantiza que el coeficiente resultante est√© siempre en el rango de ‚àí1 a 1.

**Caso del item 1**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr, kendalltau
```

```python
X = np.linspace(-10, 15, 50)

Y = (X - 3)**2 + 5

plt.scatter(X,Y)
plt.show()

r_p_scipy, p_value_pearson = pearsonr(X, Y)
print('Corr de ' + str(round(r_p_scipy,4)) + ' con p-value de ' + str(round(p_value_pearson,4)))
```

Aunque en este caso no existe una correlaci√≥n global significativa entre las variables, se observa que hay informaci√≥n valiosa. Al analizar los datos por tramos, se evidencia una correlaci√≥n negativa hasta alcanzar un m√≠nimo, y luego una correlaci√≥n positiva. Esto indica que al calcular una √∫nica correlaci√≥n para todo el conjunto de datos, se pierde informaci√≥n relevante sobre el comportamiento real de las variables.

Se observa que, al excluir el outlier, existe una correlaci√≥n entre las variables. Sin embargo, al incluirlo, la correlaci√≥n desaparece o se debilita significativamente. Esto demuestra que el coeficiente de correlaci√≥n de Pearson es sensible a la presencia de outliers.
Debido a esto el an√°lisis de correlaciones es posterior al tratamiento de la data.

**OTRA FORMA DE VER LA CORRELACION DE PEARSON, GEOMETRICAMENTE**

### Desigualdad de Cauchy-Schwarz

Para cualquier par de vectores u1,u2,...,un y v1,v1,...,vn, se cumple que:


$$
|\mathbf{u} \cdot \mathbf{v}| \leq \|\mathbf{u}\| \cdot \|\mathbf{v}\|
$$

Equivalente a la forma con sumatorias:

$$
\left( \sum_{i=1}^n u_i v_i \right)^2 \leq \left( \sum_{i=1}^n u_i^2 \right) \cdot \left( \sum_{i=1}^n v_i^2 \right)
$$

 **Recordemos la f√≥rmula de la correlaci√≥n de Pearson**:

$$
r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}
         {\sqrt{\sum (x_i - \bar{x})^2} \cdot \sqrt{\sum (y_i - \bar{y})^2}}
$$

---

### 1. Expresado como producto escalar

Sea:

-   x' = x - media(x)
-  y' = y - media(y)

Entonces:

$$
r = \frac{x' \cdot y'}{\|x'\| \cdot \|y'\|}
$$

¬°Esto es exactamente el **coseno del √°ngulo** entre los vectores centrados!

---

###  2. ¬øQu√© significa eso geom√©tricamente?

El producto escalar entre dos vectores tambi√©n se puede escribir como:

$$
x' \cdot y' = \|x'\| \cdot \|y'\| \cdot \cos(\theta)
$$

Sustituyendo en la f√≥rmula anterior:

$$
r = \cos(\theta)
$$

Donde \(\theta) es el √°ngulo entre los vectores centrados \( x' \) y \( y' \).

---

###  3. Interpretaci√≥n del valor de \( r \)

| √Ångulo \( \theta \) entre \( x' \) y \( y' \) | \( \cos(\theta) \) | Correlaci√≥n \( r \) | Interpretaci√≥n                            |

La **correlaci√≥n de Pearson** es simplemente el **coseno del √°ngulo** entre los vectores de datos centrados.

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Datos originales
x = np.array([10, 20, 30, 40, 50])
y = np.array([15, 30, 25, 50, 45])

# Paso 2: Centrado de los vectores (restar la media)
x_centered = x - np.mean(x)
y_centered = y - np.mean(y)

# Paso 3: Calcular el producto punto
dot_product = np.dot(x_centered, y_centered)

# Paso 4: Calcular las normas (magnitudes)
norm_x = np.linalg.norm(x_centered)
norm_y = np.linalg.norm(y_centered)

# Paso 5: Calcular la correlaci√≥n como coseno del √°ngulo (Cauchy-Schwarz)
r = dot_product / (norm_x * norm_y)

# Paso 6: Calcular el √°ngulo en grados
theta_rad = np.arccos(r)
theta_deg = np.degrees(theta_rad)

# Paso 7: Visualizaci√≥n de los vectores centrados (solo 2 componentes)
fig, ax = plt.subplots(figsize=(6,6))
ax.quiver(0, 0, x_centered[0], x_centered[1], angles='xy', scale_units='xy', scale=1, color='blue', label='x centrado')
ax.quiver(0, 0, y_centered[0], y_centered[1], angles='xy', scale_units='xy', scale=1, color='green', label='y centrado')

ax.set_xlim(-25, 25)
ax.set_ylim(-25, 25)
ax.set_aspect('equal')
ax.grid(True)
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.set_title("Visualizaci√≥n de vectores centrados (2 primeras componentes)")
ax.legend()
plt.show()

# Mostrar resultados
{
    "Producto punto": round(dot_product, 4),
    "Norma de x": round(norm_x, 4),
    "Norma de y": round(norm_y, 4),
    "Correlaci√≥n (r)": round(r, 4),
    "√Ångulo entre vectores (grados)": round(theta_deg, 2)
}
```

**V cramer[0,1]**
* Cualitativas o categ√≥ricas

La V de Cram√©r se define como:

$$
V = \sqrt{ \frac{\chi^2}{n \cdot (k - 1)} }
$$

Donde:
- $ \chi^2 $ es el estad√≠stico Chi-cuadrado.
- \( n \) es el total de observaciones.
- ( k = $\min(r, c)$ ), siendo \( r \) el n√∫mero de filas y \( c \) el n√∫mero de columnas de la tabla de contingencia.

0.1 < V < 0.3 d√©bil

0.3 < V < 0.5 Moderada

**Point Biseral**
* Dicotomica vs continua

$$
\rho_{pb} = \frac{M_1 - M_2}{s} \cdot \sqrt{ \frac{n_1 \cdot n_2}{n (n - 1)} }
$$

Donde:

- \( M_1 \): media de la variable continua donde la variable categ√≥rica = 1  
- \( M_2 \): media de la variable continua donde la variable categ√≥rica = 0
